```python
# Import yfinance and matplotlib
import yfinance as yf  
import matplotlib.pyplot as plt
# Get the data for the SPY ETF by specifying the stock ticker, start date, and end date
start_date = '2006-09-09'
end_date   = '2021-09-09'
```

```python
import requests
import pandas as pd

url = 'https://finance.yahoo.com/most-active?offset=0&count=25'
df = pd.read_html(requests.get(url).content)[-1]
```

```python
df.tail()
```

```python
# symbols = list(df.Symbol)

with open('symbols.txt') as f:
    symbols = f.read().splitlines() 
```

```python
symbols
```

```python
symbol_to_data = {}

for symbol in symbols:
    symbol_to_data[symbol] = yf.download(symbol, start_date, end_date).reset_index() \
    .drop(columns=['Close', 'High', 'Low', 'Adj Close'])
    symbol_to_data[symbol]['Symbol'] = symbol
```

```python
for symbol in symbols:
    print(symbol_to_data[symbol].shape)
print(f'{len(symbol_to_data)} are present')
```

```python
df = symbol_to_data[symbols[0]]

for symbol in symbols:
    df = pd.merge(df, symbol_to_data[symbol], how='outer', suffixes=[None, '_'+symbol], on='Date')
```

```python
df.head()
```

```python
f_df = df.loc[:,~df.columns.str.contains('(Volume|Symbol)')].drop(columns='Open').dropna()
```

```python
f_df = f_df.set_index('Date')
```

```python
normalized_df=(f_df-f_df.mean())/f_df.std()
```

```python
normalized_df
```

```python
def label_from_diff(row, feature_names):
    ans = []
    for feature in feature_names:
        if (row[feature] > 1e-3): ans.append(1)
        elif (row[feature] < -1e-3): ans.append(-1)
        else: ans.append(0)
    return ans
```

```python
y = normalized_df.diff().dropna() \
        .apply(lambda x: label_from_diff(x, normalized_df.columns), axis=1)
```

```python
y
```

```python
lag_to_difference = {}
with_lags = normalized_df
for i in range(1, 101):
    lag_to_difference[i] = normalized_df.diff(i)
    with_lags = pd.merge(with_lags,
                         lag_to_difference[i],
                         how='outer', 
                         suffixes=[None, f'_lag_{i}'],
                         on='Date')
```

```python
with_lags
```

```python
window_to_average = {}
with_averages = normalized_df
for i in range(5, 101):
    window_to_average[i] = normalized_df.rolling(i).mean()
    with_averages = pd.merge(with_averages,
                             window_to_average[i],
                             how='outer', 
                             suffixes=[None, f'_avg_{i}'],
                             on='Date')
```

```python
with_averages
```

```python
normalized_df
```

```python
final_df = pd.merge(with_lags, with_averages, on='Date')
```

```python
final_df
```

```python
final_df['Label'] = y
final_df = final_df.dropna()
```
