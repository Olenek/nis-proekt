```python
# Import yfinance and matplotlib
import yfinance as yf  
import matplotlib.pyplot as plt
# Get the data for the SPY ETF by specifying the stock ticker, start date, and end date
start_date = '2011-09-09'
end_date   = '2021-09-09'
```

```python
import requests
import pandas as pd

url = 'https://finance.yahoo.com/most-active?offset=0&count=25'
df = pd.read_html(requests.get(url).content)[-1]
```

```python
df.head()
```

```python
symbols = list(df.Symbol)
```

```python
symbol_to_data = {}

for symbol in symbols:
    symbol_to_data[symbol] = yf.download(symbol, start_date, end_date).reset_index() \
    .drop(columns=['Close', 'High', 'Low', 'Adj Close'])
    symbol_to_data[symbol]['Symbol'] = symbol
```

```python
len(symbol_to_data['F'])

for symbol in symbols:
    if(len(symbol_to_data['F']) > len(symbol_to_data[symbol])):
        symbols.remove(symbol)
print(f'{20 - len(symbols)} are missing')
```
